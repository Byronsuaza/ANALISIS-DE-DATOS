```python
from sklearn.model_selection import train_test_split

# Se crean variables X y Y para almacenar la informaci√≥n del entrenamiento
X = df_filtrado.drop('condition', axis=1)  # Se eliminan los datos de la columna 'condition'
Y = df_filtrado['condition']  # Se guarda la variable objetivo

# Si hay variables categ√≥ricas, se codifican (One-Hot Encoding)
X = pd.get_dummies(X, drop_first=True)

# Divisi√≥n: 75% entrenamiento y 25% prueba
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, stratify=Y, random_state=0)

# Mostrar tama√±os de los conjuntos
print("Tama√±o del conjunto de entrenamiento (X_train):", X_train.shape)
print("Tama√±o del conjunto de prueba (X_test):", X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

# Crear el modelo de regresi√≥n log√≠stica
modelo_logistico = LogisticRegression(solver='liblinear')  # 'liblinear' es eficiente para conjuntos peque√±os

# Entrenar el modelo con los datos de entrenamiento
modelo_logistico.fit(X_train, Y_train)

# Predecir con los datos de prueba
prediccion_logistica = modelo_logistico.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm_log = confusion_matrix(Y_test, prediccion_logistica)

plt.figure(figsize=(6,4))
sns.heatmap(cm_log, annot=True, fmt='d', cmap='Blues')
plt.title("Matriz de Confusi√≥n - Regresi√≥n Log√≠stica")
plt.xlabel("Predicci√≥n")
plt.ylabel("Valor Real")
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc

# Probabilidades de clase positiva
probabilidades = modelo_logistico.predict_proba(X_test)[:, 1]

# Calcular FPR, TPR y el umbral
fpr, tpr, thresholds = roc_curve(Y_test, probabilidades)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(6,4))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'√Årea bajo la curva (AUC) = {roc_auc:.2f}')
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
plt.title("Curva ROC - Regresi√≥n Log√≠stica")
plt.xlabel("Tasa de Falsos Positivos (FPR)")
plt.ylabel("Tasa de Verdaderos Positivos (TPR)")
plt.legend(loc="lower right")
plt.show()
```
